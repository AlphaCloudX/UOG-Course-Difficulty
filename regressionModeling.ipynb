Required Imports

```python
import numpy as np
from sklearn import linear_model
import pandas as pd
```

Load Data

```python
df = pd.read_csv('input_data/specificCourseInformation.csv')

x = df.iloc[1]
y = df.iloc[3]

trainData = []
target = []

for pair in zip(x[1:],y[1:]):   
    trainData.append(pair[0])
    
    target.append(pair[1])

print(trainData)
print(target)

trainingData = pd.DataFrame({'x': trainData, 'y': target})

# Save the results to a CSV file
# trainingData.to_csv('full_training_data.csv', index=False)
```

Fit regression model

Create a train, val, test split

```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

trainData = np.array(trainData).reshape(-1, 1)
target = np.array(target)

x_train, x_test, y_train, y_test = train_test_split(trainData, target, test_size=0.2, random_state=42)
```

```python
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

Train Linear Regression Model

```python
reg = linear_model.LinearRegression()

reg.fit(x_train,y_train)

# Check the model's parameters (in this case, coefficients and intercept)
print("Coefficient:", reg.coef_)
print("Intercept:", reg.intercept_)

y_pred = reg.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)

# results = pd.DataFrame({'x_test_r': x_test.flatten(), 'y_test_r': y_test, 'y_pred_r': y_pred})

# Save the results to a CSV file
# results.to_csv('model_predictions_regression.csv', index=False)
```

We see that linear regression doesn't work well so we use gradient boosting

Train A Quantile Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingRegressor

# Instantiate and fit the model
reg = GradientBoostingRegressor(loss="quantile", alpha=0.5, random_state=42)
reg.fit(x_train, y_train)

# Make predictions
y_pred = reg.predict(x_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
